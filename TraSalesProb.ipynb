```python
import numpy as np
import random
import requests
from math import dist
import gmplot
from gmplot import gmplot

api_key = "AIzaSyBmm3gvZPONdoZbwzp7xZGBkytlmaoAoEU"

coord1 = "40.080439467360904,27.960974204720902"
coord2 = "39.56570467718055,27.365627734398032"
coord3 = "39.71118609250531,28.061911547187638"
coord4 = "39.43456948716751,27.52024352878021"
coord5 = "40.07757887420205, 27.84690954853594"
coord6 = "40.020236130111215, 27.382202142601297"
coord7 = "39.72443609805689, 27.65116453874625"
coord8 = "39.72710169876507, 27.933379818680386"
coords = [coord1, coord2, coord3, coord4, coord5, coord6, coord7, coord8]

# Koordinatlar listesindeki her bir eleman için bir "for" döngüsünü çalıştırdık ve her iki nokta arasındaki mesafeyi ölçülttük. 
# Bunun için Google Maps API'si kullandık.
def get_distance(coords, api_key):
    distances = []
    for i in range(len(coords)):
        for j in range(i+1, len(coords)):
            url = f"https://maps.googleapis.com/maps/api/distancematrix/json?origins={coords[i]}&destinations={coords[j]}&mode=driving&language=tr-TR&key={api_key}"
            response = requests.get(url)
            data = response.json()
            distance = data["rows"][0]["elements"][0]["distance"]["text"]
            distances.append((f"{coords[i]} - {coords[j]}", distance))
    return distances

# Rastgele yolların bir başlangıç popülasyonunu oluşturduk.
# Popülasyon boyutunu ve koordinat listesini girdi olarak aldık, random.sample() yöntemini kullanarak,
# "pop_size" sayıda rasgele rota oluşturduk.
def create_population(pop_size, coords):
    population = []
    for i in range(pop_size):
        path = random.sample(range(len(coords)), len(coords))
        population.append(path)
    return population

#dist() işlevini kullanarak rotanın toplam mesafesini hesapladık.
#Rotadaki sırayla her bir koordinat çifti arasındaki mesafeleri toplayarak rotayı yineledik.
def get_fitness(path, coords):
    distance = 0
    for i in range(len(path)-1):
        distance += dist(coords[path[i]], coords[path[i+1]])
    return distance

#popülasyonu ve koordinat listesini parametre olarak aldık. ve get_fitness() işlevini kullanarak,
#popülasyondan toplam mesafe en düşük olanı seçtirdik.
#Her dizi bir rota ve uygunluk değerini içerdiği bir dizi liste oluşturtduk.
def selection(population, coords):
    fitness = []
    for i in population:
        fitness.append((i, get_fitness(i, coords)))
    fitness.sort(key = lambda x: x[1])
    return fitness[0][0]

#parametre olarak iki ana rotayı aldık ve çaprazlama yoluyla yeni bir alt rota oluşturmasını sağladık. 
#Ana rotalarından iki rasgele indeks (indeks1 ve indeks2) seçip ve bu indeksler arasındaki alt rotayıyı
#ana rota 1'den alt rotaya kopyalattık.
#alt rotanın kalan konumlarını, ana rota 2'den gelen ve alt rotada bulunmayan öğelerle doldurttuk.
def crossover(parent1, parent2):
    index1 = random.randint(0, len(parent1)-1)
    index2 = random.randint(0, len(parent2)-1)
    child = [-1 for i in range(len(parent1))]
    for i in range(index1, index2+1):
        child[i] = parent1[i]
    for i in range(len(parent2)):
        if parent2[i] not in child:
            for j in range(len(child)):
                if child[j] == -1:
                    child[j] = parent2[i]
                    break
    return child

#bir alt rotayı parametre alıp rotanın rastgele seçilen iki öğesini değiştirerek buna rastgele mutasyonlar getirmesini sağlattık.
def mutation(child):
    index1 = random.randint(0, len(child)-1)
    index2 = random.randint(0, len(child)-1)
    child[index1], child[index2] = child[index2], child[index1]
    return child

##Bu genetik algoritmanın ana fonksiyonudur.create_population() işlevini kullanarak rastgele yollardan oluşan başlangıç popülasyonu oluşturur. 
#Her yinelemede ana rota olarak mevcut popülasyondan en iyi yolları seçerek yeni bir popülasyon yaratıp,
#çaprazlama yoluyla yeni alt rotalar oluşturduk. Mutasyonları tanıtarak ve bunları "new_population" listesinde saklayarak,
#"num_generations" kez yinelenmesini sağladık.
def genetic_algorithm(coords, pop_size, num_generations):
    population = create_population(pop_size, coords)
    for i in range(num_generations):
        new_population = []
        for j in range(pop_size):
            parent1 = selection(population, coords)
            parent2 = selection(population, coords)
            child = crossover(parent1, parent2)
            child = mutation(child)
            new_population.append(child)
        population = new_population
    return selection(population, coords)

#Verilen koordinatlar ve API anahtarı kullanarak Google Maps API'den yol tarifi talep etmek için kullandığımız kodlar.
def get_directions(coords, api_key):
    directions = []
    for i in range(len(coords) - 1):
        url = f"https://maps.googleapis.com/maps/api/directions/json?origin={coords[i]}&destination={coords[i+1]}&mode=driving&language=tr-TR&key={api_key}"
        response = requests.get(url)
        data = response.json()
        direction = data["routes"][0]["legs"][0]["steps"]
        for step in direction:
            directions.append(step["html_instructions"])
    return directions

directions = get_directions(coords, api_key)
for direction in directions:
    print(direction)
    
    
coords = []
for coord in [coord1, coord2, coord3, coord4, coord5, coord6, coord7, coord8]:
    coord = coord.split(",")
    coords.append((float(coord[0]), float(coord[1])))

# verilen değerlere göre en kısa rota ve sürüs mesafesini yazdırıp, harita üzerinde çizdiren kodumuz.
best_path = genetic_algorithm(coords, pop_size=50, num_generations=50)
print("En kısa rota: ", best_path)
print("Sürüş mesafesi: ", get_fitness(best_path, coords))
gmap = gmplot.GoogleMapPlotter.from_geocode("Balikesir, Turkey",apikey='AIzaSyBmm3gvZPONdoZbwzp7xZGBkytlmaoAoEU')

for i in range(len(best_path)-1):
    lat1, lon1 = coords[best_path[i]]
    lat2, lon2 = coords[best_path[i+1]]
    gmap.plot([lat1, lat2], [lon1, lon2], 'cornflowerblue', edge_width=2.5)

gmap.draw("en_kisa_rota.html")
```
